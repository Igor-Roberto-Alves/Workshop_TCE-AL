```python
import pandas as pd
import json

def construir_arvore_indice(indice_raw):
    # Se o dado já for uma lista/dicionário, usa direto. Se for string JSON, decodifica.
    if isinstance(indice_raw, str):
        try:
            indice = json.loads(indice_raw)
        except:
            return {}
    else:
        indice = indice_raw

    if not indice:
        return {}

    # Raiz virtual da árvore para holding dos nós principais
    raiz = {"nome": "Raiz", "pagina": None, "regulares": [], "filhos": []}
    
    # Esta lista vai guardar o caminho (path) dos nós BOLD ativos na hierarquia
    pilha_bold = [raiz]

    for item in indice:
        nome = item.get('act')
        pagina = item.get('page')
        is_bold = item.get('bold', False)

        # ─── SUA NOVA REGRA DE TRATAMENTO DE EXCEÇÃO ───────────────────────
        # Se o texto for "Atos e Despachos", ignoramos o BOLD original e tratamos como regular
        if nome and nome.strip().lower() == "atos e despachos":
            is_bold = False
        # ───────────────────────────────────────────────────────────────────

        if is_bold:
            # Cria o novo nó BOLD
            novo_no = {
                "nome": nome,
                "pagina": pagina,
                "regulares": [],  # Filhos de mesma altura (regulares)
                "filhos": []      # Filhos BOLD subsequentes
            }
            
            # O último bold da pilha passa a ser o pai deste novo bold
            pai_atual = pilha_bold[-1]
            pai_atual["filhos"].append(novo_no)
            
            # Adiciona o novo nó à pilha, pois ele pode vir a ser pai do próximo bold
            pilha_bold.append(novo_no)
            
        else:
            # Se for regular (ou forçado a regular), pertence ao bloco BOLD ativo no momento
            no_bold_atual = pilha_bold[-1]
            no_bold_atual["regulares"].append({
                "nome": nome,
                "pagina": pagina
            })
            
            # Uma vez que encontramos um 'regular', a linha de "filhos seguidos" fecha.
            # Resetamos a pilha para a raiz:
            pilha_bold = [raiz]

    return raiz["filhos"]

# Aplicando no seu DataFrame:
df['tree'] = df['indice'].apply(construir_arvore_indice)
```

```python

import re

def buscar_termos_com_contexto(nos_da_arvore, paginas_pdf, nome_do_pai=None):
    """
    Função recursiva que varre a árvore e executa a busca na página específica.
    Exige que os nomes estejam sozinhos em suas respectivas linhas, 
    ignorando quebras de linha acidentais no meio dos nomes longos.
    """
    
    # Função auxiliar para dividir o texto e permitir quebras de linha/espaços no meio do nome
    def preparar_termo(texto):
        palavras = texto.split()
        return r'\s+'.join(re.escape(p) for p in palavras)

    for no in nos_da_arvore:
        nome_atual = no.get("nome")
        regulares = no.get("regulares", [])
        
        # Avalia cada item regular dentro deste nó (Pai/Filho)
        for index, reg in enumerate(regulares):
            reg_nome = reg.get("nome")
            num_pagina = reg.get("pagina")
            
            # Validação de segurança para garantir que a página existe na lista
            if num_pagina and num_pagina <= len(paginas_pdf):
                texto_da_pagina = paginas_pdf[num_pagina - 1]
            else:
                texto_da_pagina = "" 
            
            # REGRA 1: É o único regular neste bloco
            if len(regulares) == 1:
                termos_hierarquia = []
                
                # 1. Adiciona o Avô/Pai (se houver)
                if nome_do_pai:
                    termos_hierarquia.append(preparar_termo(nome_do_pai))
                # 2. Adiciona o Pai/Filho atual
                if nome_atual and nome_atual != nome_do_pai:
                    termos_hierarquia.append(preparar_termo(nome_atual))
                # 3. Adiciona o termo regular
                termos_hierarquia.append(preparar_termo(reg_nome))
                
                # Junta tudo com \s+
                miolo_busca = r'\s+'.join(termos_hierarquia)
                
                # ^[ \t]* garante que não há outro texto antes do início na primeira linha
                # [ \t]*$ garante que não há outro texto depois do fim na última linha
                padrao_busca = r'^[ \t]*' + miolo_busca + r'[ \t]*$'
                
                match = re.search(padrao_busca, texto_da_pagina, re.IGNORECASE | re.MULTILINE)
                
                if match:
                    print(f"✅ [ÚNICO] Achou hierarquia limpa para '{reg_nome}' na pág {num_pagina}.")
                else:
                    print(f"❌ [ÚNICO] Não achou hierarquia limpa para '{reg_nome}' na pág {num_pagina}.")
            
            # REGRA 2: Tem um irmão antes dele (índice > 0)
            elif index > 0:
                # Transforma o nome regular permitindo quebras internas, mas isolando a linha
                padrao_isolado = r'^[ \t]*' + preparar_termo(reg_nome) + r'[ \t]*$'
                
                match_isolado = re.search(padrao_isolado, texto_da_pagina, re.IGNORECASE | re.MULTILINE)
                
                if match_isolado:
                    print(f"✅ [IRMÃO] Achou isolado na linha: '{reg_nome}' na pág {num_pagina}.")
                else:
                    # Verifica se o termo existe, mas acompanhado de sujeira/outras palavras
                    if re.search(preparar_termo(reg_nome), texto_da_pagina, re.IGNORECASE):
                        print(f"⚠️ ERRO: '{reg_nome}' encontrado na pág {num_pagina}, mas NÃO está sozinho na linha!")
                    else:
                        print(f"❌ [IRMÃO] Não achou na pág {num_pagina}: '{reg_nome}'.")

        # Entra recursivamente nos nós filhos, passando o nome do nó atual como "pai"
        if no.get("filhos"):
            buscar_termos_com_contexto(no["filhos"], paginas_pdf, nome_do_pai=nome_atual)


# ─── EXECUÇÃO PARA A PRIMEIRA LINHA ──────────────────────────────

arvore_linha_0 = df["tree"].iloc[5]
paginas_linha_0 = df["pages"].iloc[5] 

print("=== INICIANDO BUSCA COM CONTEXTO ===")
buscar_termos_com_contexto(arvore_linha_0, paginas_linha_0)

```

```python
import re
import pandas as pd

def extrair_textos_dos_atos_full(arvore, texto_completo):
    """
    Varre a árvore para coletar os padrões, encontra as posições deles no texto COMPLETO (full_text)
    e fatia os blocos de texto entre um ato e outro, independentemente de quantas páginas pular.
    """
    
    # Função interna para flexibilizar a busca com Regex
    def preparar_termo(texto):
        palavras = texto.split()
        return r'\s+'.join(re.escape(p) for p in palavras)

    def obter_padroes_e_hierarquias(nos_da_arvore, nome_do_pai=None):
        padroes = []
        for no in nos_da_arvore:
            nome_atual = no.get("nome")
            regulares = no.get("regulares", [])
            
            for index, reg in enumerate(regulares):
                reg_nome = reg.get("nome")
                
                # Monta a string no formato legível: "Pai; Possível Filho; Regular"
                nomes_limpos = []
                if nome_do_pai: nomes_limpos.append(nome_do_pai)
                if nome_atual and nome_atual != nome_do_pai: nomes_limpos.append(nome_atual)
                nomes_limpos.append(reg_nome)
                hierarquia_str = "; ".join(nomes_limpos)

                # REGRA 1: É o único regular neste bloco
                if len(regulares) == 1:
                    termos_regex = []
                    if nome_do_pai: termos_regex.append(preparar_termo(nome_do_pai))
                    if nome_atual and nome_atual != nome_do_pai: termos_regex.append(preparar_termo(nome_atual))
                    termos_regex.append(preparar_termo(reg_nome))
                    
                    miolo_busca = r'\s+'.join(termos_regex)
                    padrao_busca = r'^[ \t]*' + miolo_busca + r'[ \t]*$'
                    
                    padroes.append({"hierarquia": hierarquia_str, "regex": padrao_busca})
                
                # REGRA 2: Tem um irmão antes dele (índice > 0)
                elif index > 0:
                    padrao_isolado = r'^[ \t]*' + preparar_termo(reg_nome) + r'[ \t]*$'
                    padroes.append({"hierarquia": hierarquia_str, "regex": padrao_isolado})

            # Recursão para os filhos
            if no.get("filhos"):
                padroes.extend(obter_padroes_e_hierarquias(no["filhos"], nome_atual))
                
        return padroes

    # 1. Coleta a lista de todos os padrões que precisamos procurar nesta linha
    padroes_da_linha = obter_padroes_e_hierarquias(arvore)
    
    matches_encontrados = []
    
    # 2. Faz a busca de CADA padrão no texto COMPLETO
    for padrao in padroes_da_linha:
        # Busca no texto_completo ignorando páginas
        match = re.search(padrao["regex"], texto_completo, re.IGNORECASE | re.MULTILINE)
        if match:
            matches_encontrados.append({
                "hierarquia": padrao["hierarquia"],
                "start": match.start(),
                "end": match.end()
            })
    
    # 3. Ordena os matches de acordo com a posição que aparecem no PDF inteiro (de cima para baixo)
    matches_encontrados = sorted(matches_encontrados, key=lambda x: x["start"])
    
    atos_extraidos = []
    
    # 4. Corta o texto para cada ato encontrado no documento inteiro
    for i, match in enumerate(matches_encontrados):
        inicio_texto = match["end"]
        
        # Se tiver outro ato no documento depois desse, corta até o início dele
        if i + 1 < len(matches_encontrados):
            fim_texto = matches_encontrados[i + 1]["start"]
        # Se for o ÚLTIMO ato do documento, pega tudo até a última letra do PDF
        else:
            fim_texto = len(texto_completo)
            
        texto_do_ato = texto_completo[inicio_texto:fim_texto].strip()
        
        # Adiciona a lista final no formato solicitado
        atos_extraidos.append([match["hierarquia"], texto_do_ato])
            
    return atos_extraidos


# ─── APLICAÇÃO DIRETA NO DATAFRAME ──────────────────────────────

print("Processando a coluna 'atos' usando o 'full_text' para o dataset inteiro...")

# ATENÇÃO AQUI: Mudamos de row["pages"] para row["full_text"]
df["atos"] = df.apply(lambda row: extrair_textos_dos_atos_full(row["tree"], row["full_text"]), axis=1)

print("Concluído! Verificando o resultado da primeira linha:")

primeira_linha_atos = df["atos"].iloc[0]
for ato in primeira_linha_atos:
    hierarquia = ato[0]
    # Pega os primeiros 150 caracteres para dar uma espiada se extraiu certo
    texto_resumo = ato[1][:150].replace("\n", " ") + "..." 
    print(f"\n📂 {hierarquia}\n📝 {texto_resumo}")
```
